### Problem Statement: Given an integer N, return the number of digits in N.

```python
class digits:
    def count_digits(self,n):
        count = 0
        n = abs(n)

        if n == 0:
            return 1

        while n > 0:
            count += 1
            n //= 10

        return count

if __name__ == '__main__':

    n = int(input('Enter the value of n:'))
    sol=digits()
    result = sol.count_digits(n)
    print(result)
    
```

### Reversing a number

```python
def reverse_number(n):

    reversed_num = 0
    while n > 0:
        digit = n % 10
        reversed_num = reversed_num * 10 + digit
        n //= 10
        
    return reversed_num
    
n = int(input("Enter a number: "))

print(reverse_number(n))
```

### Check whether the given number is palindrome or not!

```python
def reverse_number(n):
    original = n
    reversed_num = 0
    while n > 0:
        digit = n % 10
        reversed_num = reversed_num * 10 + digit
        n //= 10
        
        
    return reversed_num == original
    
n = int(input("Enter a number: "))
print(reverse_number(n))
```

### Checking for an Armstrong number!

```python
def is_armstrong(n):
    original = n
    digits = len(str(n))
    total = 0

    while n > 0:
        digit = n % 10
        total += digit ** digits
        n //= 10

    return total == original


n = int(input("Enter a number: "))

if is_armstrong(n):
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")
```

### Printing all the divisors of N

```python
def divisors(n):
    divisor = []
    i = 1

    while i * i <= n:
        if n % i == 0:
            divisor.append(i)

            if i != n//i:
                divisor.append(n//i)

        i += 1

    divisor.sort()

    for d in divisor:
        print(d, end = ' ')

        
n=int(input("Enter the number:"))
result = divisors(n)
result 
```

### Checking for a prime number

```python
def is_prime(n):
    if n <=1:
        return False

    i = 2
    while i * i <= n:
        if n % i == 0:
            return False
        i += 1

    return True

n = int(input('Enter the number:'))

if is_prime(n):
    print('It is a Prime Number!')

else:
    print('It is not a Prime Number!')
```
