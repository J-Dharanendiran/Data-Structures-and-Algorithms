```python
class solution:
    def pattern(self,n):
        for i in range(1,n+1):

            for j in range(n-i): # For printing the leading space
                print(' ',end=' ')

            for j in range(2*i-1):
                print('*',end=' ')

            print()

if __name__ == '__main__' :
    n=int(input('Enter the number of rows'))
    sol=solution()
    sol.pattern(n)
```

```python
class solution:
    def pattern(self,n):
        for i in range(n):
            # Upper half
            # Leading space
            for j in range(n-i-1):
                print(' ',end =' ')
            # Printing star
            for j in range(2*i+1):
                print('*', end=' ')
            print()
        # Lower half
        for i in range(n,0,-1):
            #Leading space
            for j in range(n-i):
                print(' ', end = ' ')
            #Printing star
            for j in range(2*i-1):
                print("*", end=' ')
            print()

if __name__ == "__main__":
    n=int(input('Enter the no of rows:'))
    sol=solution()
    sol.pattern(n)
    
```

```python
class Solution:
    def pattern(self, n):

        for i in range(1,n+1):

            if i % 2 == 0:
                start = 1
            else:
                start = 0

            for j in range(i):
                print(start, end=" ")

                start = 1 - start

            print()


if __name__ == "__main__":
    n = int(input("Enter n: "))

    sol = Solution()
    sol.pattern(n)
```

```python
class solution:
    def pattern(self,n):
        for i in range(1,n+1):
            #numbers
            for j in range(1,i+1):
                print(j,end='')
            #inbetween space
            for j in range(2*(n-i),0,-1):
                print(' ',end='')
            #numbers
            for j in range(i,0,-1):
                print(j,end='')

            print()

if __name__ == '__main__':
    n=int(input("Enter the no of rows:"))
    
    sol=solution()
    sol.pattern(n)
```
